```python
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer

words=['running','happiness','fishing','easily','studies','cries','nationality']
porter=PorterStemmer()


porter_steems=[porter.stem(word) for word in words]
print("original -> porter_stems")
for i,w in enumerate(words):
  print(f'{w:12}'f'->{porter_steems[i]:8}')

  fig,ax=plt.subplots(figsize=(10,5))
  bar_width=0.25
  x=range(len(words))
  ax.bar(words,porter_steems)
```

```python
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

words=['write','writing','writes','wrote','writer']
ps=PorterStemmer()
ls=LancasterStemmer()
ss=SnowballStemmer('english')

poster_steams=[ps.stem(w) for w in words]
lancaster_steams=[ls.stem(w) for w in words]
snowball_steams=[ss.stem(w) for w in words]

for i,w in enumerate(words):
    print(f"{w:12}->{poster_steams[i]:8} | {lancaster_steams[i]:9} | {snowball_steams[i]}")

fig,ax=plt.subplots(figsize=(10,6))

bar_width = 0.25
x = range(len(words))

ax.bar([i-bar_width/2 for i in x],[len(s) for s in poster_steams],width=bar_width,label='Porter-steams')
ax.bar(x,[len(s) for s in lancaster_steams],width=bar_width,label='Lancaster_steams')
ax.bar([i+bar_width/2 for i in x],[len(s) for s in snowball_steams],width=bar_width,label='Snowball_steams')
```

```python
import nltk
from nltk import pos_tag,word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tree import Tree
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text="the quick brown fox jumps over the lazy dog near the river bank"
tokens=word_tokenize(text)
tagged_tokens=pos_tag(tokens)
print(tagged_tokens)
chunk_grammar=r"NP:{<DT>?<JJ>*<NN>}"
chunk_parser=nltk.RegexpParser(chunk_grammar)
tree=chunk_parser.parse(tagged_tokens)
tree.pretty_print()
```
